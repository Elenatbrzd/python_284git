```python
df = pd.read_csv("nhanes.csv")
```

```python
#Example 1 Current-smokers & BMI>25
df.query("SmokingStatus=='Current' & BMI>25")
```

```python
#Example 2 90th percentile of BMI
df.BMI.quantile(0.9)
print(df.BMI.quantile(0.9))
df[df.BMI > df.BMI.quantile(0.9)]
```

```python
#Example 3
df[["id","Height","Weight","BMI"]].sort_values(by="BMI", ascending=False)
```

```python
#Example 4 
Education=df.groupby("Education")
Education["Income"].mean()
```

```python
#Example 5
df.query("Gender == 'female' & Age>18").sort_values(by="Pulse", ascending = False).head(10)
```

```python
#Example 6
df.groupby("Race").BPSys.describe()
```

```python
#Example 7
Active= df[["id","Education","PhysActive","PhysActiveDays"]]
Active.assign(PCTActive = Active.PhysActiveDays*100/7).sort_values(by="PCTActive").head(10)
```

```python
#Example 8
A=df.query("Income<Income.quantile(0.5)").groupby("Race")
A["Weight"].mean()
```
